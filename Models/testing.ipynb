```python
from transformers import GPT2TokenizerFast, GPT2LMHeadModel
```

```python

tokenizer = GPT2TokenizerFast.from_pretrained('nehalelkaref/poem_generator')
poetry_generator = GPT2LMHeadModel.from_pretrained('nehalelkaref/poem_generator')
```

```python
test_sentences = [ 'وكم أرض جدب دونها ولصوص',
                  'كأني ورحلي والقراب ونمرقي',
                  'كفدح النضي باليدين المفوق'
                  ]
```

```python
import torch
if torch.cuda.is_available():
    device = torch.device('cuda')
else:
    print('No GPU available, using CPU')
    device = torch.device('cpu')
        
poetry_generator.cuda() 
```

```python
prompt = tokenizer.bos_token + test_sentences[0] + tokenizer.eos_token
encoded_prompt = tokenizer.encode(prompt, return_tensors = 'pt').to(device)
gneration = poetry_generator.generate(encoded_prompt, top_k = 70, max_length =50, top_p= 0.2,
                           repetition_penalty = 3.0,
    no_repeat_ngram_size = 3, num_beams=15)
                   
generated_text = tokenizer.decode(gneration[0], skip_special_tokens = True)
print(generated_text)
```
